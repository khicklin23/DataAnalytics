```python
#6.a monthly report for year 2017-2018, which has monthly average open price, close price, 
# transaction volume and gain/loss, and a query to find all of the months which have certain range of open prices
data['Date'] = pd.to_datetime(data['Date'])
data["Gain"]=data['Close']-data['Open']
Months_2017= data[data["Date"].dt.year==2017].groupby(data["Date"].dt.month).mean()
Months_2017.index.name="Month"
data["Gain"]=data['Close']-data['Open']
Months_2017= Months_2017.loc[:, ["Open", "Close", "Volume", "Gain"]]
print(Months_2017)

def SearchOpen(i, j, data):
  return data[data["Open"]<j][data["Open"]>i]

SearchOpen(2300, 2500, Months_2017)
```

```python
#7.
Years = data.groupby(data["Date"].dt.year).mean()
Years.index.name="Year"
Years = Years.loc[:, ["Open", "Close", "Volume", "Gain"]]
print(Years)

Year_Dates = data.groupby(data["Date"].dt.year).agg({'Date' :['min', 'max']})
Year_Dates.columns = Year_Dates.columns.droplevel(0)
Year_Dates.reset_index(inplace=True)

Year_Dates = pd.merge(Year_Dates, data[['Date', 'Open']], left_on=["min"], right_on=["Date"])
Year_Dates = pd.merge(Year_Dates, data[['Date', 'Close']], left_on=["max"], right_on=["Date"])
Year_Dates["Profit"]=Year_Dates["Close"]-Year_Dates["Open"]
print(Year_Dates.loc[:, ['Date_x','Profit']])

print(Year_Dates.loc[Year_Dates['Profit']==Year_Dates['Profit'].max(), ['Date_x', 'Profit']])
```

### Using what we have learned in Unit5 of Pandas to analyze Sacramentorealestatetransactions.csv file with Series and DataFrame operations. Please find answers to the following questions:

1. Regroup the data first by city name, then by type

2. For each city, each type, find the highest, median and lowest transactions

3. For each zipcode and each type, find average transactions

```python
#1.
#load data
import pandas as pd
data = pd.read_csv("Sacramentorealestatetransactions.csv")

#grouping by city then by type
data2 = data.groupby(['city', 'type'])

#view the groups
for name, group_data in data2:
    print(f"Name: {name[0]}\nType: {name[1]}")
    print(group_data)
    print("\n---------------------------------------------------------------------------------\n")
```

```python
#2.
#grouping by city then by type
data2 = data.groupby(['city', 'type'])

#view the groups
for city, group_data in data2:
    print(f"Name: {city[0]}\nType: {city[1]}\n")
    print("Max Price:")
    print(group_data.price.max())
    print("Median Price:")
    print(group_data.price.median())
    print("Min Price:")
    print(group_data.price.min())
    print("\n---------------------------------------------------------------------------------\n")
```

```python
#3. 
data.groupby(['zip', 'type']).agg({'price': ['mean']})
```
